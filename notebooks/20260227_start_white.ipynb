```python
import pandas as pd
```

```python
df = pd.read_csv('../data_out/20260227_215902.csv')
df = df.drop(columns=["position_fen"])
df.head()
```

```python
df.columns
```

```python
df["diff"] = (
    df["source_lichess_user_white_percent"] - df["source_lichess_db_white_percent"]
)
df["total"] = df["source_lichess_user_total"] + df["source_chesscom_user_total"]
df["perc"] = df["total"] / df["total"].sum() * 100
df["cum_perc"] = df["perc"].cumsum()
df["pot"] = df["diff"] * df["perc"]
df["pot_cum"] = df["pot"].where(df["pot"] < 0, 0).cumsum()

df[
    [
        "move_san",
        "move_eval_cp",
        "source_lichess_user_total",
        "source_lichess_user_white_percent",
        "source_lichess_db_white_percent",
        "diff",
        "total",
        "perc",
        "cum_perc",
        "pot",
        "pot_cum",
    ]
]
```
